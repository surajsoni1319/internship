# 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get('https://www.imdb.com/chart/top/?ref_=nv_mv_250')
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
```

```python
movies_name = soup.find_all('td', class_='titleColumn')
movies_name
```

```python
movies = []
for movie in movies_name:
    movie = movie.get_text().replace('\n', "")
    movie = movie.strip(" ")
    movies.append(movie)
movies
```

```python
#IMDB rating 
movies_rating = soup.find_all('td', class_='ratingColumn imdbRating')
movies_rating
```

```python
ratings = []
for rating in movies_rating:
    rating = rating.get_text().replace('\n', '')
    ratings.append(rating)
ratings
```

```python
data = pd.DataFrame()
data['movies'] = movies
data['rating'] = ratings
data.head(100)
```

# 7) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world :

i) Headline
ii) Time
iii) News Link

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page=requests.get("https://www.cnbc.com/world/?region=world")
page
```

```python
soup = BeautifulSoup(page.content)
```

```python
#Headline

headline = []
for i in soup.find_all('a',class_="LatestNews-headline"):
    headline.append(i.text)
    
headline
```

```python
#Time 

time = []
for i in soup.find_all('time',class_="LatestNews-timestamp"):
    time.append(i.text)
    
time
```

```python
url = []
for i in soup.find_all("a",class_="LatestNews-headline"):

    url.append(i.get("href"))
    
url
```

```python
# Data Frame

df = pd.DataFrame({"HEADLINE":headline, "TIME":time,"URL":url})
df
```

# 9) Write a python program to scrape mentioned details from dineout.co.in : 

    i) Restaurant name
    ii) Cuisine
    iii) Location
    iv) Ratings
    v) Image URL

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page=requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
```

```python
# i) Restaurant name
titles = []
for i in soup.find_all('div',class_="restnt-info cursor"):
    titles.append(i.text)
    
titles
```

```python
# ii) Cuisine

cuisine = []
for i in soup.find_all('div',class_="detail-info"):
    cuisine.append(i.text.split('|')[1])
    
cuisine
```

```python
# iii) Location

Location = []
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    Location.append(i.text)
    
Location
```

```python
# iv) Ratings

rating = []
for i in soup.find_all('div',class_="rating-txt hide"):
    rating.append(i.text)
    
rating
```

```python
# v) Image URL

url = []
for i in soup.find_all('img',class_="no-img"):
    url.append(i['data-src'])
    
url
```

```python
df = pd.DataFrame({"Restaurant name":titles, "Cuisine":cuisine, "Location":Location, "Ratings":rating, "URL":url})
df
```

# 10) Write a python program to scrape the details of top publications from Google Scholar from https://scholar.google.com/citations?view_op=top_venues&hl=en

i) Rank
ii) Publication
iii) h5-index
iv) h5-median

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page=requests.get("https://scholar.google.com/citations?view_op=top_venues&hl=en")
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
```

```python
# rank
rank = []
for i in soup.find_all('td',class_="gsc_mvt_p"):
    rank.append(i.text)
    
rank
```

```python
#publication 
pub = []
for i in soup.find_all('td',class_="gsc_mvt_t"):
    pub.append(i.text)
    
pub
```

```python
#h5 Index

h5_index = []
for i in soup.find_all('a',class_="gs_ibl gsc_mp_anchor"):
    h5_index.append(i.text)
    
h5_index
```

```python
#H5 Median

h5_median = []
for i in soup.find_all('span',class_="gs_ibl gsc_mp_anchor"):
    h5_median.append(i.text)
    
h5_median
```

```python
print(len(rank),len(pub),len(h5_index),len(h5_median))
```

```python
df = pd.DataFrame({"rank":rank, "publication":pub, "h5_index":h5_index,"h5_median":h5_median},index=rank)
df  
```
